# Solar home electricity data by Ausgrid company
# https://www.ausgrid.com.au/Industry/Our-Research/Data-to-share/Solar-home-electricity-data

```python
# 2012-2013 Solar home electricity data v2.csv

import pandas as pd
# read the veteran dataset
df_raw = pd.read_csv('2012-2013 Solar home electricity data v2.csv')
df_raw.info()
df_raw.head(20)
```

##### 2012-2013 Solar home electricity data... = customer ID
##### Unnamed: 1 = Generator Capacity
##### Unnamed: 2 = Postcode
##### Unnamed: 3 = Consumption Category	
##### Unnamed: 4 = date
##### Unnamed 5 ~ 52 = 0:30 - 0:00

```python
df_raw.columns = ['Customer','Generator Capacity','Postcode','Consumption Category','date','0:30',
                  '1:00','1:30','2:00','2:30','3:00','3:30','4:00','4:30','5:00',
                  '5:30','6:00','6:30','7:00','7:30','8:00','8:30','9:00','9:30',
                  '10:00','10:30','11:00','11:30','12:00','12:30','13:00','13:30',
                  '14:00','14:30','15:00','15:30','16:00','16:30','17:00','17:30',
                  '18:00','18:30','19:00','19:30','20:00','20:30','21:00','21:30',
                  '22:00','22:30','23:00','23:30','0:00','Row Quality']
```

```python
df_raw.info()
df_raw.head(5)
```

## Drop frist row, every first value is column name

```python
df = df_raw.iloc[1:]
df
```

```python
df.info()
```

```python
df['Customer'] = df['Customer'].astype(str)
```

```python
df['Customer'].unique()
```

```python
mask = df['Customer'] == '299'
data = pd.DataFrame(df[mask])
data.info()
data
```

##### GC = General Consumption for electricity supplied all the time (primary tariff, either inclining block or time of use rates),
##### excluding solar generation and controlled load supply
##### CL = Controlled Load Consumption (Off peak 1 or 2 tariffs)
##### GG = Gross Generation for electricity generated by the solar system with a gross metering configuration, measured separately
##### to household loads

##### GC+CL = whole energy consumption of a day, CL is not necessary for every family


```python
import numpy as np
import matplotlib.pyplot as plt
np.mean(data['0:30'])
```

```python
plt.figure(figsize = (30, 8))
plt.plot(data['0:30'], color = 'green', label = "Produce")
```

```python
GC = data.iloc[0, 6:-1]
```

```python
GG = data.iloc[1, 6:-1]
```

```python
plt.figure(figsize = (30, 8))
plt.plot(GG, color = 'green', label = "Produce")
plt.plot(GC, color = 'Red', label = "Consumption")
plt.legend()
```

```python
total = GG-GC
```

```python
plt.figure(figsize = (30, 8))
plt.plot(total, color = 'blue', label = "Total")
plt.legend()
```
