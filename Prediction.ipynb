```python
def find_different_values(array1, array2):
    # Initialize an empty list to store the different values
    different_values = []

    # Loop through the elements in the first array
    for item in array1:
        # If the element is not in the second array, add it to the different_values list
        if item not in array2:
            different_values.append(item)

    # Loop through the elements in the second array
    for item in array2:
        # If the element is not in the first array and not already in different_values, add it
        if item not in array1 and item not in different_values:
            different_values.append(item)

    return different_values

noclCustomer = find_different_values(clCustomer, allCustomer)
# print(noclCustomer)
noclCustomer = noclCustomer[1:]
print(len(noclCustomer))
```
